```python
import os
import re
import subprocess
from collections import Counter
from googleapiclient.discovery import build

VIDEO_ID = "2Gg6Seob5Mg"  # Correct video ID

youtube = build("youtube", "v3", developerKey=API_KEY)

def extract_comments(video_id, max_comments=100):
    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=max_comments,
        textFormat="plainText"
    )

    response = request.execute()

    for item in response["items"]:
        comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
        comments.append(comment)

    return comments

def extract_timestamps(comments):
    timestamps = []
    pattern = r"\b(\d{1,2}:\d{2})\b"  # Matches timestamps like 12:34 or 5:45

    for comment in comments:
        found = re.findall(pattern, comment)
        timestamps.extend(found)

    return Counter(timestamps).most_common(5)  # Get top 5 most mentioned timestamps

def download_video(video_url, output_filename="video.mp4"):
    """Download YouTube video using yt-dlp."""
    command = [
        "yt-dlp",
        "-f", "best",  # Best quality available
        "-o", output_filename,
        video_url
    ]
    subprocess.run(command, check=True)
    print(f"Downloaded video as {output_filename}")

def resize_video(input_file, output_file, width=720, height=1280):
    """Resize video using ffmpeg."""
    command = [
        "ffmpeg",
        "-i", input_file,
        "-vf", f"scale={width}:{height}",
        "-c:a", "copy",
        output_file
    ]
    subprocess.run(command, check=True)
    print(f"Resized video saved as {output_file}")

# Extract most mentioned timestamps
comments = extract_comments(VIDEO_ID)
popular_timestamps = extract_timestamps(comments)

print("Most mentioned timestamps in comments:")
for time, count in popular_timestamps:
    print(f"{time} - mentioned {count} times")

# Download and resize video
video_url = f"https://www.youtube.com/watch?v={VIDEO_ID}"
download_video(video_url, "original.mp4")
resize_video("original.mp4", "resized.mp4", 720, 1280)
```
